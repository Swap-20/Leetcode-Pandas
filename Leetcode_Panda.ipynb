#### 1873 - Calculate Special Bonus

Calculate the bonus of each employee. The bonus of an employee is 100% of their salary if the ID of the employee is an odd number and the employee's name does not start with the character 'M'. The bonus of an employee is 0 otherwise.

Return the result table ordered by employee_id.

```python
data = [[2, 'Meir', 3000], [3, 'Michael', 3800], [7, 'Addilyn', 7400], [8, 'Juan', 6100], [9, 'Kannon', 7700]]
employees = pd.DataFrame(data, columns=['employee_id', 'name', 'salary']).astype({'employee_id':'int64', 'name':'object', 'salary':'int64'})
employees
```

```python
df = employees[employees['name'].str.contains("M")  & employees['employee_id'] % 2 != 0]
df
```

```python
employees['bonus'] = employees.apply(
    lambda x:x['salary'] if x['employee_id'] % 2  and not x['name'].startswith('M') else 0,
    axis = 1
)
df = employees[['employee_id','bonus']].sort_values('employee_id')
df
```
